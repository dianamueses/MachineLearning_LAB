# 0.) Import and Clean data

```python
import pandas as pd
from google.colab import drive
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn.preprocessing import StandardScaler

import seaborn as sns
from sklearn.decomposition import PCA
```

```python
drive.mount('/content/gdrive/', force_remount = True)
```

```python
df = pd.read_csv("/content/gdrive/MyDrive/ColabTrial/Country-data.csv", sep = ",")
```

```python
df.head()
```

```python
df.columns
```

```python
names = df[["country"]]
X = df.drop(["country"], axis = 1)
```

```python
scaler = StandardScaler().fit(X)
X_scaled = scaler.transform(X)
```

# 1.) Run a PCA Algorithm to get 2 Principle Components for the 9 X features

```python
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca
```

# 2.) Plot a Scatter plot of the PCs on the axis

```python
plt.subplots(figsize = (15, 5), dpi = 300)
plt.scatter(x = X_pca[:,0], y = X_pca[:,1], color = 'plum')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Plot for Xs Reduced')
plt.plot()
```

As shown on the graph above, our Xs have kept their variance, even after reducing dimensionality and multicolinearities.

# 3.) Rank the features in order of importance according to PCA

```python
loadings = pca.components_
```

```python
# PCA 1 and 2
loadings
```

```python
# feature importance for first 9 variables
feature_importance = pd.DataFrame(np.sum(loadings**2, axis = 0))
feature_names = df.columns[1:]
feature_importance.index = feature_names
feature_importance.sort_values(0, ascending = False)
```

# 4.) Plot a heatmap of the feature importance (Fill in all parameters)

```python
#sns.heatmap(, annot=, cmap='coolwarm', xticklabels=, yticklabels=[, ], cbar_kws={'label':''})
sns.heatmap(loadings, annot = True, cmap='coolwarm', xticklabels = feature_names, 
            yticklabels=['PC1','PC2'], cbar_kws={'label':'Feature Importance in PCA'})
plt.xlabel('Original Features')
plt.ylabel('Principal Components')
plt.title('Loadings Heatmap')
plt.show()
```

# 5.) Plot a correlation plot of the original features. What do you notice between the graphs of 4 & 5?

```python
sns.heatmap(X.corr(), annot = True)
plt.plot()
```

As we can see on heatmap 4, child_mort and total_fer are the least important variables for PC1, while exports and imports are the most important for PC2. This can be confirmed by looking at correlation plot 5, in which total_fer and child_mort seem to be highly negatively correlated, while exports and imports are highly positively correlated.

# 6.) Run a PCA with 9 PCs. Plot a Cumulative Explained Variance Plot. How many PCs should we use if we want to retain 95% of the variance?

```python
pca = PCA(n_components = 9)
X_pca = pca.fit_transform(X_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
cumulative_explained_variance = np.cumsum(pca.explained_variance_ratio_)
plt.subplots(figsize = (15, 5), dpi = 300)
plt.plot(np.arange(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', color = 'plum')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance Plot')
plt.grid()
plt.show()
```

```python
cumulative_explained_variance
```

From our results, we know that at least 6 PCAs are required if we want to retain 95% of the variance.
